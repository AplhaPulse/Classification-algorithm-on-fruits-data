```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from matplotlib.colors import ListedColormap
```

```python
Feature_Names = ['Mass', 'Width', 'Height', 'Color_Score']
X = Fruits[Feature_Names]
Fruit_Label = 'Fruit_Label'
y = Fruits['Fruit_Label']
```

```python
cmap = ListedColormap(['#FF0000', '#00FF00', '#0000FF', '#FFFF00'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split(X,y,random_state = 0)
```

```python

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)



```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)

logreg_accuracy_train = logreg.score(X_train, y_train)
logreg_accuracy_test = logreg.score(X_test, y_test)

print(f'Accuracy of Logistic Regression classifier on training set: {logreg_accuracy_train:.2f}')
print(f'Accuracy of Logistic Regression classifier on test set: {logreg_accuracy_test:.2f}')
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

clf_accuracy_train = clf.score(X_train, y_train)
clf_accuracy_test = clf.score(X_test, y_test)

print(f'Accuracy of DecisionTreeClassifier on Training set:{clf_accuracy_train:.2f}')
print(f'Accuracy of DecisionTreeClassifier on test set:{clf_accuracy_test:.2f}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

knn_accuracy_train = knn.score(X_train, y_train)
knn_accuracy_test = knn.score(X_test, y_test)

print(f'Accuracy of KNeighborsClassifier on Training set:{knn_accuracy_train:.2f}')
print(f'Accuracy of KNeighborsClassifier on test set:{knn_accuracy_test:.2f}')


```

```python
from sklearn.discriminant_analysis  import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)

lda_accuracy_train = lda.score(X_train, y_train)
lda_accuracy_test = lda.score(X_test, y_test)

print(f'Accuracy of LinearDiscriminantAnalysis on Training set: {lda_accuracy_train:.2f}')
print(f'Accuracy of LinearDiscriminantAnalysis on test set: {lda_accuracy_test:.2f}')



```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train, y_train)

gnb_accuracy_train = gnb.score(X_train, y_train)
gnb_accuracy_test = gnb.score(X_test, y_test)

print(f'Accuracy of Gaussian Naive Bayes on Training set: {gnb_accuracy_train:.2f}')
print(f'Accuracy of Gaussian Naive Bayes on test set: {gnb_accuracy_test:.2f}')

```

```python
from sklearn.svm import SVC

svm = SVC()
svm.fit(X_train, y_train)

svm_accuracy_train = svm.score(X_train, y_train)
svm_accuracy_test = svm.score(X_test, y_test)

print(f'Accuracy of SVM on Training set: {svm_accuracy_train:.2f}')
print(f'Accuracy of SVM on test set: {svm_accuracy_test:.2f}')
```

```python
#KNN algorithm
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
pred = knn.predict(X_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```
